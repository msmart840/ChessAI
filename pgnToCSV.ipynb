```python
def moves_array(moves):
    moves = moves[6:]
    end_str = moves.find('...')
    moves = moves[0:end_str]
    moves_array = moves.split(' ')
    return moves_array

def play_game(moves):
    board = chess.Board()
    board_data = []
    n_moves = len(moves)
    for i in range(n_moves):
        board.push_san(moves[i])
        x = board_to_array(board)
        board_data.append(board_to_array(x))
        print(i)
    return board_data

pieces = ['p','r','n','b','k','q','P','R','N','B','K','Q','.']


def board_to_array(board):
    board_array = [0]*(65)
   
    for i in range(64):
        piece = str(board.piece_at(i))
        board_array[i] = piece
        
    board_array[-1] = int(board.turn)
    board_array = [np.asarray(board_array)]
    return board_array
```

```python
n_games = 0
max_games = 100000
for filepath in glob.iglob('data/pgns/*.pgn'):
    
    gm = 1
    try : 
        pgn = open(filepath)         
        print(filepath)
        X = []
        Y = []
        Game_N = []
        move_n_list = []
        move_to_end = []
        for i in pgn:
            game = chess.pgn.read_game(pgn)
            move_n = 0
            moves = []

            if game != None:
                board = game.board()
                board_new = []
                for move in game.mainline_moves():
                    board.push(move)
                    board_new.append(board_to_array(board))
                    moves.append(move_n)
                    move_n += 1

                moves_fin = []
                for i in range(len(moves)):
                    moves_fin.append(move_n - moves[i])

                move_n_list = move_n_list  + moves
                move_to_end = move_to_end  + moves_fin
                Game_N = Game_N + [n_games]*len(board_new)
                result_str = game.headers["Result"]
                if result_str == '1/2-1/2': result = "Draw"
                if result_str == '0-1': result = "Black"
                if result_str == '1-0': result = "White"
                #rint(result_str + ' ---- ' + str(result))
                results_new = [result]*len(board_new)
                X = X + board_new
                Y = Y + results_new
                #print('GAME NUMBER ' + str(gm) + ' ' + str(game.headers['White']) + ' VS ' + str(game.headers['Black']))
                gm +=1 
                n_games +=1
        X1 = np.concatenate(X,axis=0)

        data = pd.DataFrame({
            "result" : Y,
            "move" : move_n_list,
            "moveEnd" : move_to_end, 
            "game" : Game_N

            })

        for i in range(65) : 
            data["X" + str(i)] = X1[:,i]
        name = pgn.name[10:-4].replace("/","").replace(".","")  
        data.to_csv("data/" + name + ".csv")
    except :
        print("error with PGN")
```
